```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import datetime
import glob
```

```python
# # Raw URLs of your CSV files
January = "https://raw.githubusercontent.com/Sempi23/CitiBike-Tableau/main/2024_tripdata/JC-202401-citibike-tripdata.csv"
Febuary = "https://raw.githubusercontent.com/Sempi23/CitiBike-Tableau/main/2024_tripdata/JC-202402-citibike-tripdata.csv"
March = "https://raw.githubusercontent.com/Sempi23/CitiBike-Tableau/main/2024_tripdata/JC-202403-citibike-tripdata.csv"
April = "https://raw.githubusercontent.com/Sempi23/CitiBike-Tableau/main/2024_tripdata/JC-202404-citibike-tripdata.csv"

# Get all CSV filenames within the directory
filenames = [January, Febuary, March, April]

# Loop through filenames and read data
dfs = []
for file in filenames:
    print(f'Reading: {file}')
    dfs.append(pd.read_csv(file))
```

```python
# Loop through the DataFrames and print the first few rows of each DataFrame for a quick review
for df in dfs:
    display(df.head(2))  # Print the first few rows of each DataFrame
```

```python
# Get information about the columns. Confirm that the data type for the two variables is now dateime64
for df in dfs:
    display(df.info()) 
```

```python
# Concatenate the DataFrames in the dfs list
concatenated_df = pd.concat(dfs)

# Display information about the concatenated DataFrame
print(concatenated_df.info())
```

```python
# Number of individual bikes
concatenated_df['ride_id'].unique().shape
```

```python
#Number of citibike stations
concatenated_df['start_station_id'].unique().shape
```

```python
# Drop null
concatenated_df.dropna(inplace=True)
```

```python
concatenated_df.shape
```

```python
#Convert Columns to DateTime Fomat
concatenated_df['started_at'] = pd.to_datetime(concatenated_df['started_at'])
concatenated_df['ended_at'] = pd.to_datetime(concatenated_df['ended_at'])
```

```python
# Create colum for day of the week
concatenated_df['day_of_week'] = concatenated_df['started_at'].dt.day_name()
```

```python
# Caldulate the duration of each ride
# Convert datetime columns to numpy datetime64
start_time = pd.to_datetime(concatenated_df['started_at'])
end_time = pd.to_datetime(concatenated_df['ended_at'])

# Calculate trip duration in seconds using timedelta64
trip_duration_seconds = (end_time - start_time) / np.timedelta64(1, 's')

# Convert trip duration to minutes and round off to 2 decimal points
trip_duration_minutes = round(trip_duration_seconds / 60, 2)

# Add trip duration in minutes to the DataFrame
concatenated_df['trip_duration_minutes'] = trip_duration_minutes
```

```python
# Drop any rows with less than 2 minutes trip duration
concatenated_df = concatenated_df[concatenated_df['trip_duration_minutes'] >= 2]
concatenated_df.shape
```

```python
#Calculate the distance between two points.An efficient way to calculate the distance between two points on the Earth's surface is to use the Haversine formula.
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of the Earth in kilometers

    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    distance_km = R * c
    distance_miles = distance_km * 0.621371  # Convert kilometers to miles
    return distance_miles

# Calculate distance in miles for each row in the DataFrame
concatenated_df['distance_miles'] = haversine(concatenated_df['start_lat'],
                                               concatenated_df['start_lng'],
                                               concatenated_df['end_lat'],
                                               concatenated_df['end_lng'])

# Round distance in miles to 2 decimal points
concatenated_df['distance_miles'] = concatenated_df['distance_miles'].round(2)

#Preview data
concatenated_df.head()
```
